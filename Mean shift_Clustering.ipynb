```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x = dataset.iloc[:,3:5].values
```

```python
from sklearn.cluster import MeanShift

ms = MeanShift(bandwidth=2, bin_seeding=True)

ms.fit(x)
```

```python
cluster_centers = ms.cluster_centers_
cluster_centers.shape
```

```python
y_pred = ms.fit_predict(x)
y_pred
```

```python
Supervised = dataset
```

```python
Supervised["Cluster_Group"] = y_pred
Supervised
```

```python
import seaborn as sns
graph = sns.lmplot(data=Supervised, x=Supervised.columns[3],y=Supervised.columns[4], hue = Supervised.columns[5], 
          fit_reg=True,legend=True,legend_out=True)
```
